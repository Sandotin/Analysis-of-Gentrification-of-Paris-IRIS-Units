# GRAPHS SHOWING PROPORTION AMOUNG PARIS POPULATION

POPULATION PROPORTION BASED ON POPULATION DEMOGRAPHY

```python
IRIS_DEMOGRAPHIE = Table.read_table('IRIS_DEMOGRAPHIE-2.csv')
```

```python
IRIS_DEMOGRAPHIE = IRIS_DEMOGRAPHIE.where('NOM_EPT', "Paris").where('P12_POP',are.above_or_equal_to(0))
```

```python
IRIS_DEMOGRAPHIE = IRIS_DEMOGRAPHIE.select('NOM_EPT','P12_POP','P12_POP_ETR','P12_POP_IMM','C12_POP15P_CS1','C12_POP15P_CS2','C12_POP15P_CS3','C12_POP15P_CS4','C12_POP15P_CS5','C12_POP15P_CS6')
```

```python
IRIS_DEMOGRAPHIE = IRIS_DEMOGRAPHIE.group('NOM_EPT',sum) 
```

```python
IRIS_DEMOGRAPHIE = Table().with_columns('Categories', make_array('Population in 2012','Number of foreigners',
'Number of Immigrants','Number of people between the age of 15 and more, who are farmer',
'Number of people between the age 15 years and over, who are Craftsmen, Traders, Entrepreneurs',
'Number of people aged 15 and over who are Senior managers and have higher intellectual occupations',
'Number of people 15 years and over who have an Intermediate intellectual occupations',
'Number of people between 15 years and over who are Employees (who have a job)',
'Number of people aged 15 and over factory workers (more as in factory workers)'),
                                           'Values', np.array(IRIS_DEMOGRAPHIE.drop(0).row(0)))
```

```python
IRIS_DEMOGRAPHIE.take([0,1,2]).barh('Categories')
```

```python
IRIS_DEMOGRAPHIE.take([3,4,5,6]).barh('Categories')
```

```python
IRIS_DEMOGRAPHIE.take([7,8]).barh('Categories')
```

POPULATION PROPORTION BASED ON POPULATION SOCIOLOGY

```python
IRIS_SOCIOLOGIE = Table.read_table('IRIS_SOCIOLOGIE.csv')
```

```python
IRIS_SOCIOLOGIE = IRIS_SOCIOLOGIE.where('NOM_EPT', "Paris").where('C12_MEN',are.above_or_equal_to(0))
```

```python
IRIS_SOCIOLOGIE = IRIS_SOCIOLOGIE.select('NOM_EPT','P12_RP_VOIT1P','C12_MEN')
```

```python
IRIS_SOCIOLOGIE = Table().with_columns('Categories', make_array('Number of households with at least one car',
                                                                'Number of households'),
                                           'Values', np.array(IRIS_SOCIOLOGIE.drop(0).row(0)))
```

```python
IRIS_SOCIOLOGIE.barh('Categories')
```

POPULATION PROPORTION BASED ON POPULATION ACTIVIES

```python
IRIS_ACTIVITE = Table.read_table('IRIS_ACTIVITE.csv')
```

```python
IRIS_ACTIVITE = IRIS_ACTIVITE.where('NOM_EPT', "Paris").where('C12_ACTOCC15P',are.above_or_equal_to(0))
```

```python
IRIS_ACTIVITE= IRIS_ACTIVITE.select('NOM_EPT','P12_ACT1564','P12_ACTOCC1564',
                                    'C12_ACTOCC15P','C12_ACTOCC15P_TCOM','C12_ACT1564_CS1',
                                    'C12_ACT1564_CS2','C12_ACT1564_CS3','C12_ACT1564_CS4',
                                    'C12_ACT1564_CS5','C12_ACT1564_CS6')
```

```python
IRIS_ACTIVITE = IRIS_ACTIVITE.group('NOM_EPT',sum)
```

```python
IRIS_ACTIVITE = Table().with_columns('Categories', make_array('Number of active people aged 15 to 64', 
                                                              'Number of employed persons aged 15 to 64',
'Number of employed persons aged 15 or over',
'Number of employed persons 15 years of age and over who primarily use public transportation to work',
'Number of active farmers aged 15 to 64',
'Number of craftsmen, shopkeepers, active entrepreneurs aged 15 to 64',
'Number of senior managers and higher intellectual occupations active aged 15 to 64',
'Number of active intermediate occupations aged 15 to 64',
'Number of active employees aged 15 to 64',
'Number of active workers (factory workers) aged 15 to 64'),
                                           'Values', np.array(IRIS_ACTIVITE.drop(0).row(0)))
```

```python
IRIS_ACTIVITE.take([0,1]).barh('Categories')
```
